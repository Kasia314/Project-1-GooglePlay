# 3. Hypothesis testing

I will test two hypothesis:

1:	average price of paid ame app is 4$ 

2: checking if average size of games within different versions was the same 


```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns 
%matplotlib inline

from scipy.stats import ttest_1samp
from scipy.stats import norm,t, sem

sns.set()

pd.set_option('max_columns', 50)
data = pd.read_csv('Clean_games.csv')
df = data.copy()
df.sample(10)
```

## 1. I will test the hypothesis that average price for paid (not free) game is 4$.

 Ho: df.Price.mean = 4$
 
 H1: ~above

```python
# extracting paid apps
paid_games = df.loc[(df['Price']!=0)]
paid_games
```

```python
# checking for price outliers
df.loc[(df['Price']!=0)].describe()
```

```python
# I see there are right side outliers which might make my distribution not N, I will visualize it

sns.set()
plt.figure(figsize = (14,10))
ax = sns.distplot(df.loc[(df['Price']!=0)]['Price'])
```

```python
# I decide to get rid of Price outliers

summary=df.loc[(df['Price'] != 0)].describe().T
summary.head()
```

```python
summary['IQR']=summary['75%']-summary['25%']
summary.head()

summary['cutoff']=summary.IQR*1.7
summary.head()

summary['lw']=summary['25%']-summary.cutoff
summary['rw']=summary['75%']+summary.cutoff
summary.head()
```

```python
#extraction paid games without Price outliers

paid_games_wo_outliers = paid_games.loc[(paid_games.Price > 0) & (paid_games.Price < 7.79)]
```

```python
# I see there are right side outliers which might make my distribution not N, I will visualize it

sns.set()
plt.figure(figsize = (14,10))
ax = sns.distplot(paid_games_wo_outliers['Price'])
```

```python
# It is still not N distributed, so I shouldnt make a ttest, but just for a sake of practice, I will ignore the not Normal distribution of Price
# I will not present these results in a final presentation
paid_games_wo_outliers.Price.mean()
```

```python
ttest_1samp(paid_games_wo_outliers.Price,4)
```

```python
# pvalue < 0.05, therefore, we cannot reject Ho that 4$ is an average price for an App
```

```python
# prconfidence intervals:
t.interval(0.95, df=82, loc=paid_games_wo_outliers.Price.mean(), scale=sem(paid_games_wo_outliers.Price))
```

```python
# Probability that real Price mean (from the population) will fall within 2.42 and 3.23 is 95%.
```

## 2. I will take two samples of game apps from two different Android Versions and, using T-test for the means of two independent samples, I will check if the average size off games were the same in both samples.

Ho:	avg of Sample1.Size = avg of Sample2.Size 

H1: avg of Sample1.Size =! avg of Sample2.Size 

```python
import numpy as np
from scipy.stats import ttest_ind, ttest_ind_from_stats
from scipy.special import stdtr


sample1 = df.loc[(df['New_version'] == '4.1')]
sample2 = df.loc[(df['New_version'] == '4.0.3')]


sample1 = sample1[sample1['New_size'].notna()]
sample2 = sample2[sample2['New_size'].notna()]

# Use scipy.stats.ttest_ind.
t, p = ttest_ind(sample1['New_size'], sample2['New_size'], equal_var=False)
print("ttest_ind:            t = %g  p = %g" % (t, p))
```

```python
# pvalue < 0,05 
# therefore, we can reject Ho that average Size in Game category for 4.0.3 version is equal to those for 4.1 version.

# We can explain it by the way that app developpers are well following the expanded possibilities that new Android versions give 
```

```python
# I will check the distribution of both samples

plt.figure(figsize = (6,3))
ax = sns.distplot(sample1['New_size'])
ax.set_xlabel('App size in KB')
ax.set_title('App ver. 4.1 Size Distribution')
plt.show()

plt.figure(figsize = (6,3))
ax = sns.distplot(sample2['New_size'])
ax.set_xlabel('App size in KB')
ax.set_title('App ver. 4.0.3 Size Distribution')
plt.show()
```

```python
# it is kind of Normaal...
```

```python
#I will check for other Android versions:

sample1 = df.loc[(df['New_version'] == '4.0') & (df['Category'] == 'GAME')]
sample2 = df.loc[(df['New_version'] == '5.0') & (df['Category'] == 'GAME')]


sample1 = sample1[sample1['New_size'].notna()]
sample2 = sample2[sample2['New_size'].notna()]

# Use scipy.stats.ttest_ind.
#t, p = ttest_ind(sample1.Size, sample2.Size, equal_var=False)
t, p = ttest_ind(sample1['New_size'], sample2['New_size'], equal_var=False)
print("ttest_ind:            t = %g  p = %g" % (t, p))
```
