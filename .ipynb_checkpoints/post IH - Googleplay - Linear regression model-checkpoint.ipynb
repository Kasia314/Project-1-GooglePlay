## 5. Linear regression model

 - Predicting number of installations of game app basing on other features.

```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns 
%matplotlib inline

from statsmodels.api import OLS
from statsmodels.formula.api import ols
from statsmodels.api import add_constant


pd.set_option('max_columns', 50)
data = pd.read_csv('Clean_games.csv')
df = data.copy()
df.sample(10)
```

```python
# checking the correlation of all apps
cleanapps = pd.read_csv('Clean_GoogleApps.csv')
plt.figure(figsize=(12, 12))
sns.heatmap(cleanapps.corr(), annot=True, square=True,cmap="YlGnBu")
```

```python
# the most evident and very big linear correlation is seen between installs and reviews
```

# Model 1

```python
df.columns
```

```python
# I will try to predict Installs_log basing on other features

Y = df.Installs_log
X = df[['Reviews_log', 'Arcade', 'Card', 'Puzzle','Racing']]

model = OLS(Y,add_constant(X))
fitted_model = model.fit()
fitted_model.summary()
```

```python
# after playing with a lot of features, my best result is the one above with R-squared 0,93, AIC =3373 and BIC =3403.

# adding predictions to my table (logs and normal)

predictions = fitted_model.predict()
predictions
```

```python
df['Predicted_log']=fitted_model.predict()
df['Predicted_normal'] = df['Predicted_log'].apply(lambda x: np.exp(x))
df
```

```python
sns.regplot(x="Installs_log", y=predictions, data=df)
```

```python
#CHECKING ASSUMPTIONS:

import Assumptions as ass

tester=ass.Assumption_Tester_OLS(X,Y)
tester.run_all()
```
