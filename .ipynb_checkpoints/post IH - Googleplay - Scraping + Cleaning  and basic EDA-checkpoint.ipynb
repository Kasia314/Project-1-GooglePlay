## 4. Web scraping and EDA of scraped data

```python
#####  Scraping List of Android smartphones with Android version and release date  #####
    
import requests as r
from bs4 import BeautifulSoup
import time
import json
import re
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_Android_smartphones'
    
soup=BeautifulSoup(r.get(url).content)
table = soup.select('.wikitable')[0]
ll=[[i.text.strip('\n\xa0') for i in row.find_all(['td','th'])] for row in table.find_all('tr')]
phones = pd.DataFrame(ll[1:],columns=ll[0])
phones

    
```

```python
phones.to_csv(("phones.csv"), index=False)
```

```python
phones['Android version at release'].value_counts()
```

```python
phones['Android ver clean'] =phones['Android version at release'].apply(lambda x: x.strip('Android '))
phones["Android ver clean"].value_counts()
```

```python
df1=phones.groupby(by='Android ver clean')[['Model']].count().sort_values(by='Model')
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df1.plot.bar(figsize=(10,8))
```

```python
df2=phones.groupby(by='Developer')[['Model']].count().sort_values(by='Model', ascending=False).head(10)
df2
```

```python
df2.plot.bar(figsize=(10,8))
```

```python
df4=phones.groupby(['Release year','Developer'])[['Model']].count().sort_values(by='Release year', ascending=False).head(20)
df4
```

```python
df4.plot.bar(figsize=(10,8))
```

```python
#phones['Release date'].apply(lambda x: x.split('/'))

year_lst = []
for row in phones['Release date']:
    year = ''.join(re.findall('[0-9]{4}',row))   
    year_lst.append(year)
    
# adding a column Month:
phones['Release year'] = year_lst

phones
```

```python
phones.to_csv(("phones.csv"), index=False)
```

```python
#3. How many Android supported phones were released each year

df3=phones.groupby(by='Release year')[['Model']].count().sort_values(by='Release year').head(20)
df3
```

```python
df3.plot.bar(figsize=(10,8))
```
